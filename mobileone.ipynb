```python
# ### Step 1: Install and Import Libraries ###
# First, make sure you have all the necessary libraries installed. 
# If not, you can install them by running this command in your terminal:
# pip install torch torchvision timm datasets scikit-learn seaborn matplotlib captum

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from timm.models import create_model
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_dataset
from captum.attr import LayerGradCam
from captum.attr import visualization as viz
import os


# ### Step 2: Define Focal Loss ###
# Here we implement the Focal Loss function, which is effective for 
# datasets with significant class imbalance.
class FocalLoss(nn.Module):
    """Focal Loss, as described in https://arxiv.org/abs/1708.02002.
    
    It is designed to address class imbalance by down-weighting
    easy examples and focusing on hard examples.
    """
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.alpha is not None:
            # Ensure alpha is on the same device as the loss
            alpha_t = self.alpha[targets].to(focal_loss.device)
            focal_loss = alpha_t * focal_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# ### Step 3: Data Loading and Preprocessing ###
# We load the `marmal88/skin_cancer` dataset, create a mapping from 
# string labels to integers, and set up PyTorch Datasets and DataLoaders 
# with appropriate data augmentation.

# Load dataset
ds = load_dataset("marmal88/skin_cancer")

# Create label mapping from dx strings to integers
classes = ds['train'].unique('dx')
class_to_idx = {cls: i for i, cls in enumerate(classes)}
idx_to_class = {i: cls for cls, i in class_to_idx.items()}
print(f"Classes found: {classes}")
print(f"Class to index mapping: {class_to_idx}")

# PyTorch Dataset Class
class SkinDataset(Dataset):
    def __init__(self, split, transform=None):
        self.dataset = ds[split]
        self.transform = transform
        
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        item = self.dataset[idx]
        image = item['image']
        # Convert string label to integer using the mapping
        label = class_to_idx[item['dx']]
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.long)

# Data augmentation and preprocessing
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    normalize
])

val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    normalize
])

# Create datasets
train_dataset = SkinDataset('train', transform=train_transform)
val_dataset = SkinDataset('validation', transform=val_test_transform)
test_dataset = SkinDataset('test', transform=val_test_transform)

# Check class distribution and compute weights for Focal Loss alpha
train_labels = [class_to_idx[label] for label in ds['train']['dx']]
class_weights = compute_class_weight(
    class_weight='balanced', 
    classes=np.unique(train_labels),
    y=train_labels
)
class_weights = torch.tensor(class_weights, dtype=torch.float32)
print("Class weights for loss function:", class_weights)

# --- OPTIMIZATION FOR GTX 1650 ---
# Reduced batch size to prevent CUDA out-of-memory errors on a 4GB GPU.
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=0, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=8, num_workers=0, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=8, num_workers=0, pin_memory=True)


# ### Step 4: Model Configuration and Training Setup ###
# We initialize the `mobileone_s0.apple_in1k` model, set up the Focal Loss, 
# AdamW optimizer, and the Cosine Annealing with Warm Restarts scheduler.

# Setup device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Initialize model
model = create_model('mobileone_s0.apple_in1k', pretrained=True, num_classes=len(classes))
model = model.to(device)

# Loss function, optimizer, and scheduler
criterion = FocalLoss(alpha=class_weights, gamma=2.0).to(device)
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-2)
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=1, eta_min=1e-6)

# Use Automatic Mixed Precision (AMP) for faster training and less memory usage
scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())


# ### Step 5: Training and Validation Loop ###
# Define the functions for a single training epoch and for validation. Then, 
# run the main training loop for 50 epochs, saving the best model based on validation loss.

# --- OPTIMIZATION FOR GTX 1650 ---
# Gradient accumulation steps. Effective batch size will be (batch_size * accumulation_steps) = 16 * 2 = 32
ACCUMULATION_STEPS = 2

# Training loop with Gradient Accumulation
def train_epoch(model, loader, criterion, optimizer, scaler, device, accumulation_steps):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    # Reset gradients at the start of the epoch
    optimizer.zero_grad()

    for i, (images, labels) in enumerate(loader):
        images, labels = images.to(device), labels.to(device)
        
        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            outputs = model(images)
            loss = criterion(outputs, labels)
            # Normalize loss to account for accumulation
            loss = loss / accumulation_steps
        
        # Accumulate gradients
        scaler.scale(loss).backward()
        
        # --- OPTIMIZATION FOR GTX 1650 ---
        # Update weights only every `accumulation_steps`
        if (i + 1) % accumulation_steps == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad() # Reset gradients for the next accumulation cycle
        
        running_loss += (loss.item() * accumulation_steps) * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Validation loop
def validate(model, loader, criterion, device):
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
    epoch_loss = val_loss / len(loader.dataset)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Main training execution
num_epochs = 50
best_val_loss = float('inf')
history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

print("\n--- Starting Training ---")
for epoch in range(num_epochs):
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, scaler, device, ACCUMULATION_STEPS)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    
    scheduler.step() # Step the scheduler every epoch
    
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_acc'].append(train_acc)
    history['val_acc'].append(val_acc)
    
    # This print statement now explicitly shows all the requested metrics for each epoch
    print(f"Epoch {epoch+1}/{num_epochs}: "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
          f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f} | "
          f"LR: {optimizer.param_groups[0]['lr']:.6f}")
    
    # Save the best model
    if val_loss < best_val_loss:
        torch.save(model.state_dict(), 'best_mobileone_s0_model.pth')
        best_val_loss = val_loss
        print(f"   -> Best model saved with val_loss: {best_val_loss:.4f}")
print("--- Training Finished ---")


# ### Step 6: Evaluation and Visualization ###
# After training, we load the best model and evaluate its performance on the test set. 
# We then generate a classification report, a confusion matrix, and plots for 
# the training/validation loss and accuracy curves.

# Load best model for evaluation
model.load_state_dict(torch.load('best_mobileone_s0_model.pth'))
test_loss, test_acc = validate(model, test_loader, criterion, device)
print(f"\nTest Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")

# Generate classification report
y_true, y_pred = [], []
model.eval()
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

class_names = list(classes)
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# Plot confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('confusion_matrix.png')
plt.show()

# Plot training and validation curves
plt.figure(figsize=(15, 5))

# Plot for Loss
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Loss Curves (Train vs. Validation)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot for Accuracy
plt.subplot(1, 2, 2)
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['val_acc'], label='Validation Accuracy')
plt.title('Accuracy Curves (Train vs. Validation)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('training_curves.png')
plt.show()


# ### Step 7: Grad-CAM Visualization with Captum ###
# We use `LayerGradCam` from the `captum` library to visualize which parts of an image 
# the model is focusing on for its prediction. This is a powerful technique for model interpretability.
# Note: Finding the correct `target_layer` is model-specific. For `mobileone_s0`, 
# a good choice is `model.stages[-1].blocks[-1].conv_pw`.

def visualize_grad_cam(model, loader, device, target_layer, idx_to_class):
    model.eval()
    images, labels = next(iter(loader))
    img_tensor = images[0:1].to(device) # Take a single image
    true_label = labels[0].item()

    # Initialize LayerGradCam
    layer_gc = LayerGradCam(model, target_layer)
    
    # Get prediction
    with torch.no_grad():
        output = model(img_tensor)
        pred_label = torch.argmax(output, dim=1).item()

    # Compute attributions
    attribution = layer_gc.attribute(img_tensor, target=pred_label)
    
    # Prepare original image for visualization
    original_image = np.transpose(img_tensor.squeeze(0).cpu().detach().numpy(), (1, 2, 0))
    # Un-normalize
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    original_image = (original_image * std) + mean
    original_image = np.clip(original_image, 0, 1)

    # Visualize
    fig, _ = viz.visualize_image_attr(
        np.transpose(attribution.squeeze(0).cpu().detach().numpy(), (1, 2, 0)),
        original_image,
        method="blended_heat_map",
        sign="all",
        show_colorbar=True,
        title=f"Predicted: {idx_to_class[pred_label]} | True: {idx_to_class[true_label]}"
    )
    fig.savefig('grad_cam.png')
    print("Grad-CAM visualization saved as 'grad_cam.png'")

# Define the target layer for Grad-CAM
# This layer choice is specific to the 'mobileone_s0' architecture.
target_layer_mobileone = model.stages[-1].blocks[-1].conv_pw

# Generate and save visualization
try:
    visualize_grad_cam(model, test_loader, device, target_layer_mobileone, idx_to_class)
except Exception as e:
    print(f"Could not generate Grad-CAM visualization. Error: {e}")

print("\n--- Script finished successfully! ---")

```

```python
# ### Step 1: Install and Import Libraries ###
# First, make sure you have all the necessary libraries installed. 
# If not, you can install them by running this command in your terminal:
# pip install torch torchvision timm datasets scikit-learn seaborn matplotlib captum

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
from timm.models import create_model
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_dataset
from captum.attr import LayerGradCam
from captum.attr import visualization as viz
import os

# --- FIX: Define device early to be used globally ---
# Setup device (CUDA or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")


# ### Step 2: Define Focal Loss ###
# Here we implement the Focal Loss function, which is effective for 
# datasets with significant class imbalance.
class FocalLoss(nn.Module):
    """Focal Loss, as described in https://arxiv.org/abs/1708.02002.
    
    It is designed to address class imbalance by down-weighting
    easy examples and focusing on hard examples.
    """
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss

        if self.alpha is not None:
            # Ensure alpha is on the same device as the loss
            alpha_t = self.alpha[targets].to(focal_loss.device)
            focal_loss = alpha_t * focal_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# ### Step 3: Data Loading and Preprocessing ###
# We load the `marmal88/skin_cancer` dataset, create a mapping from 
# string labels to integers, and set up PyTorch Datasets and DataLoaders 
# with appropriate data augmentation.

# Load dataset
ds = load_dataset("marmal88/skin_cancer")

# Create label mapping from dx strings to integers
classes = ds['train'].unique('dx')
class_to_idx = {cls: i for i, cls in enumerate(classes)}
idx_to_class = {i: cls for cls, i in class_to_idx.items()}
print(f"Classes found: {classes}")
print(f"Class to index mapping: {class_to_idx}")

# PyTorch Dataset Class
class SkinDataset(Dataset):
    def __init__(self, split, transform=None):
        self.dataset = ds[split]
        self.transform = transform
        
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        item = self.dataset[idx]
        image = item['image']
        # Convert string label to integer using the mapping
        label = class_to_idx[item['dx']]
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.long)

# Data augmentation and preprocessing
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    normalize
])

val_test_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    normalize
])

# Create datasets
train_dataset = SkinDataset('train', transform=train_transform)
val_dataset = SkinDataset('validation', transform=val_test_transform)
test_dataset = SkinDataset('test', transform=val_test_transform)

# Check class distribution and compute weights for Focal Loss alpha
train_labels = [class_to_idx[label] for label in ds['train']['dx']]
class_weights = compute_class_weight(
    class_weight='balanced', 
    classes=np.unique(train_labels),
    y=train_labels
)
# --- FIX: Move class_weights tensor to the correct device ---
class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
print("Class weights for loss function:", class_weights)

# --- OPTIMIZATION FOR GTX 1650 ---
# Reduced batch size to prevent CUDA out-of-memory errors on a 4GB GPU.
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=0, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=8, num_workers=0, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=8, num_workers=0, pin_memory=True)


# ### Step 4: Model Configuration and Training Setup ###
# We initialize the `mobileone_s0.apple_in1k` model, set up the Focal Loss, 
# AdamW optimizer, and the Cosine Annealing with Warm Restarts scheduler.

# Initialize model
model = create_model('mobileone_s0.apple_in1k', pretrained=True, num_classes=len(classes))
model = model.to(device)

# Loss function, optimizer, and scheduler
criterion = FocalLoss(alpha=class_weights, gamma=2.0).to(device)
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-2)
scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=1, eta_min=1e-6)

# Use Automatic Mixed Precision (AMP) for faster training and less memory usage
scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())


# ### Step 5: Training and Validation Loop ###
# Define the functions for a single training epoch and for validation. Then, 
# run the main training loop for 50 epochs, saving the best model based on validation loss.

# --- OPTIMIZATION FOR GTX 1650 ---
# Gradient accumulation steps. Effective batch size will be (batch_size * accumulation_steps) = 16 * 2 = 32
ACCUMULATION_STEPS = 2

# Training loop with Gradient Accumulation
def train_epoch(model, loader, criterion, optimizer, scaler, device, accumulation_steps):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    # Reset gradients at the start of the epoch
    optimizer.zero_grad()

    for i, (images, labels) in enumerate(loader):
        images, labels = images.to(device), labels.to(device)
        
        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            outputs = model(images)
            loss = criterion(outputs, labels)
            # Normalize loss to account for accumulation
            loss = loss / accumulation_steps
        
        # Accumulate gradients
        scaler.scale(loss).backward()
        
        # --- OPTIMIZATION FOR GTX 1650 ---
        # Update weights only every `accumulation_steps`
        if (i + 1) % accumulation_steps == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad() # Reset gradients for the next accumulation cycle
        
        running_loss += (loss.item() * accumulation_steps) * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
    epoch_loss = running_loss / len(loader.dataset)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Validation loop
def validate(model, loader, criterion, device):
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
    epoch_loss = val_loss / len(loader.dataset)
    epoch_acc = correct / total
    return epoch_loss, epoch_acc

# Main training execution
num_epochs = 50
best_val_loss = float('inf')
history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

print("\n--- Starting Training ---")
for epoch in range(num_epochs):
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, scaler, device, ACCUMULATION_STEPS)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    
    scheduler.step() # Step the scheduler every epoch
    
    history['train_loss'].append(train_loss)
    history['val_loss'].append(val_loss)
    history['train_acc'].append(train_acc)
    history['val_acc'].append(val_acc)
    
    # This print statement now explicitly shows all the requested metrics for each epoch
    print(f"Epoch {epoch+1}/{num_epochs}: "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
          f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f} | "
          f"LR: {optimizer.param_groups[0]['lr']:.6f}")
    
    # Save the best model
    if val_loss < best_val_loss:
        torch.save(model.state_dict(), 'best_mobileone_s0_model.pth')
        best_val_loss = val_loss
        print(f"   -> Best model saved with val_loss: {best_val_loss:.4f}")
print("--- Training Finished ---")


# ### Step 6: Evaluation and Visualization ###
# After training, we load the best model and evaluate its performance on the test set. 
# We then generate a classification report, a confusion matrix, and plots for 
# the training/validation loss and accuracy curves.

# Load best model for evaluation
model.load_state_dict(torch.load('best_mobileone_s0_model.pth'))
test_loss, test_acc = validate(model, test_loader, criterion, device)
print(f"\nTest Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")

# Generate classification report
y_true, y_pred = [], []
model.eval()
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

class_names = list(classes)
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# Plot confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('confusion_matrix.png')
plt.show()

# Plot training and validation curves
plt.figure(figsize=(15, 5))

# Plot for Loss
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Loss Curves (Train vs. Validation)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot for Accuracy
plt.subplot(1, 2, 2)
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['val_acc'], label='Validation Accuracy')
plt.title('Accuracy Curves (Train vs. Validation)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('training_curves.png')
plt.show()


# ### Step 7: Grad-CAM Visualization with Captum ###
# We use `LayerGradCam` from the `captum` library to visualize which parts of an image 
# the model is focusing on for its prediction. This is a powerful technique for model interpretability.
# Note: Finding the correct `target_layer` is model-specific. For `mobileone_s0`, 
# a good choice is `model.stages[-1].blocks[-1].conv_pw`.

def visualize_grad_cam(model, loader, device, target_layer, idx_to_class):
    model.eval()
    images, labels = next(iter(loader))
    img_tensor = images[0:1].to(device) # Take a single image
    true_label = labels[0].item()

    # Initialize LayerGradCam
    layer_gc = LayerGradCam(model, target_layer)
    
    # Get prediction
    with torch.no_grad():
        output = model(img_tensor)
        pred_label = torch.argmax(output, dim=1).item()

    # Compute attributions
    attribution = layer_gc.attribute(img_tensor, target=pred_label)
    
    # Prepare original image for visualization
    original_image = np.transpose(img_tensor.squeeze(0).cpu().detach().numpy(), (1, 2, 0))
    # Un-normalize
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    original_image = (original_image * std) + mean
    original_image = np.clip(original_image, 0, 1)

    # Visualize
    fig, _ = viz.visualize_image_attr(
        np.transpose(attribution.squeeze(0).cpu().detach().numpy(), (1, 2, 0)),
        original_image,
        method="blended_heat_map",
        sign="all",
        show_colorbar=True,
        title=f"Predicted: {idx_to_class[pred_label]} | True: {idx_to_class[true_label]}"
    )
    fig.savefig('grad_cam.png')
    print("Grad-CAM visualization saved as 'grad_cam.png'")

# Define the target layer for Grad-CAM
# This layer choice is specific to the 'mobileone_s0' architecture.
target_layer_mobileone = model.stages[-1].blocks[-1].conv_pw

# Generate and save visualization
try:
    visualize_grad_cam(model, test_loader, device, target_layer_mobileone, idx_to_class)
except Exception as e:
    print(f"Could not generate Grad-CAM visualization. Error: {e}")

print("\n--- Script finished successfully! ---")

```
